```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

raw = pd.read_csv('T2-Group1 - missingdata (1).csv')
                  
raw
```

```python
print('There are',len(raw),'records.') #print rows
```

```python
print('There are',len(raw.iloc[0]),'columns.') #print columns
```

```python
Nraw = raw[raw['Condition'] ==  "Normal1"] #inner condition returns T/F for each row, then saves T rows in new df

C3raw = raw[raw['Condition'] ==  "Condition3"] 

C4raw = raw[raw['Condition'] ==  "Condition4"]
```

```python
print('There are',len(Nraw),'records in Normal 1 condtion')
print('There are',len(C3raw),'records in Condition 3 condtion')
print('There are',len(C4raw),'records in Condition 4 condtion')
```

```python
#df.isnull()                                                  labels every cell as T if it is null, F if otherwise
#df.sum()                                                     sums every entry for each column
#df.isnull().sum()                                            sums every null for each column
#df.isnull().sum().sum()                                      sums the sum from every row -> counts all nulls
print('Null count breakdowns for:\n', sep='\n')

print('Normal:' , Nraw.isnull().sum() , sep='\n') #by default, print() appends a ' ' after each 'text' segment
print('\nCondition 3:' , C3raw.isnull().sum() , sep='\n') #sep='\n' replaces that with a new line instead for readability
print('\nCondition 4:' , C4raw.isnull().sum() , sep='\n')
```

```python
#df.iloc[[0,-1]].isnull().sum().sum()     ->      sums the total number of null values in the first and last rows of df

print('Number of null values in the first and last rows of the different conditions are:' , sep='\n')
print('Normal :',Nraw.iloc[[0,-1]].isnull().sum().sum())
print('Condition 3 :',C3raw.iloc[[0,-1]].isnull().sum().sum())
print('Condition 4 :',C4raw.iloc[[0,-1]].isnull().sum().sum())

# Since the null values are not found in the first or last records of the 3 different condtions, we can account forward and backward fill
# forward fill, backward fill
```

```python
nullrecordsN = []                                              # finds all the nulls in df and stores it in
nullrecordsC3 = []                                                  # a list of lists nullrecords
nullrecordsC4 = []

for col in range(len(Nraw.iloc[0])):                       
    for record in range(len(Nraw)):
        if pd.isnull(Nraw.iloc[record][col]) == True:
            nullrecordsN.append([record,col])
            
for col in range(len(C3raw.iloc[0])):                       
    for record in range(len(C3raw)):
        if pd.isnull(C3raw.iloc[record][col]) == True:
            nullrecordsC3.append([record,col])
            
for col in range(len(C4raw.iloc[0])):                       
    for record in range(len(C4raw)):
        if pd.isnull(C4raw.iloc[record][col]) == True:
            nullrecordsC4.append([record,col])
```

```python
#works because nullrecords is built in ascending order of column of nan values, not records of nan values
for i in range(len(nullrecordsN)-1):
    if (nullrecordsN[i][1] == nullrecordsN[i+1][1]) and (nullrecordsN[i][0]+1 == nullrecordsN[i+1][0]):
        print('Consecutive nulls detected in Nraw')

for i in range(len(nullrecordsC3)-1):
    if (nullrecordsC3[i][1] == nullrecordsC3[i+1][1]) and (nullrecordsC3[i][0]+1 == nullrecordsC3[i+1][0]):
        print('Consecutive nulls detected in C3raw')

for i in range(len(nullrecordsC4)-1):
    if (nullrecordsC4[i][1] == nullrecordsC4[i+1][1]) and (nullrecordsC4[i][0]+1 == nullrecordsC4[i+1][0]):
        print('Consecutive nulls detected in C4raw')
        
#testrecords = [[1,1],[4,1],[4,2],[3,5],[4,5]]           FOR TESTING. The last 2 coords will trigger print function
                                                         # Again, the list must be built in ascending column values           
#for i in range(len(testrecords)-1):
    #if (testrecords[i][1] == testrecords[i+1][1]) and (testrecords[i][0]+1 == testrecords[i+1][0]):
        #print('Consecutive nulls detected')
```

```python
# The last 3 lines check for nulls in /1: first and last rows\ and /2: consecutive records in the same column\
# Since no errors were raised, we can safely interpolate/backwardfill/forwardfill safely
# At this point, data checking is complete.
# We can now proceed to data preprocessing (filling up the null values via interpolate)
```

```python
# Using pandas' built in function to fill null values with interpolation
Nfilled = Nraw.interpolate()
C3filled = C3raw.interpolate()
C4filled = C4raw.interpolate()
```

```python
dd=pd.concat([Nfilled,C4filled,C3filled]) # concatenate the 3 dataframes after filling in the null values
dd
```

```python
dd=dd[["Condition","Cylinder heating zone 1", "Cylinder heating zone 2", "Cylinder heating zone 3", "Cylinder heating zone 4", "Cylinder heating zone 5", "Mould temperature control unit 1", "Maximum injection pressure", "Cycle time", "Injection time", "Dosage time", "Switch-over volume", "Material cushion"]]
# adjust to coordinate cylinder heating zone timing
```

```python
dd['Cylinder heating zone 1']=dd['Cylinder heating zone 1'].round(2) #round all the parameters for easier calculation in statistical analysis
dd['Cylinder heating zone 2']=dd['Cylinder heating zone 2'].round(2)
dd['Cylinder heating zone 3']=dd['Cylinder heating zone 3'].round(2)
dd['Cylinder heating zone 4']=dd['Cylinder heating zone 4'].round(2)
dd['Cylinder heating zone 5']=dd['Cylinder heating zone 5'].round(2)
dd['Mould temperature control unit 1']=dd['Mould temperature control unit 1'].round(2)
dd['Maximum injection pressure']=dd['Maximum injection pressure'].round(2)
dd['Cycle time']=dd['Cycle time'].round(3)
dd['Injection time']=dd['Injection time'].round(3)
dd['Dosage time']=dd['Dosage time'].round(3)
dd['Switch-over volume']=dd['Switch-over volume'].round(5)
dd['Material cushion']=dd['Material cushion'].round(6)
```

```python
dd.to_csv(r'Proj1ReorderedRound.csv', index=False) # export new csv after preprocessing to be used for visualisations
```
